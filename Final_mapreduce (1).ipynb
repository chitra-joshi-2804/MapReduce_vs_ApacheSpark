```python
df = df.dropna()
df.info()
```

```python
import time

# Start time for MapReduce Model
start_time = time.time()
```

```python
# Feature selection
features = ['Open', 'High', 'Low', 'Volume']
X = df[features]
y = df['Close']
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.model_selection import cross_val_score
# Define the map function to generate individual predictions (map step)
def map_function(chunk):
    model = RandomForestRegressor(n_estimators=100, random_state=1)
    model.fit(chunk[0], chunk[1])
    predictions = model.predict(X_test)
    return predictions
```

```python
from sklearn.model_selection import cross_val_score
# Split the data into smaller chunks for map function (simulating MapReduce)
def split_data(X_train, y_train, n_chunks):
    chunk_size = len(X_train) // n_chunks
    X_chunks = [X_train[i:i + chunk_size] for i in range(0, len(X_train), chunk_size)]
    y_chunks = [y_train[i:i + chunk_size] for i in range(0, len(y_train), chunk_size)]
    #print(X_chunks)
    #print(y_chunks)
    return list(zip(X_chunks, y_chunks))
```

```python
# Map step: Create multiple predictions using different chunks of data
n_chunks = 4
chunks = split_data(X_train, y_train, n_chunks)

mapped_predictions = list(map(map_function, chunks))
```

```python
# Reduce function: Average the predictions from different models (reduce step)
def reduce_function(mapped_predictions):
    return np.mean(mapped_predictions, axis=0)
```

```python
# Final predictions after reducing
final_predictions = reduce_function(mapped_predictions)
```

```python
# Calculate the performance of the model using Mean Squared Error
mse = mean_squared_error(y_test, final_predictions)
print(f"Mean Squared Error of the model: {mse}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Calculate RMSE
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Calculate MAE
mae = mean_absolute_error(y_test, final_predictions)
print(f"Mean Absolute Error (MAE): {mae}")

# Calculate R-squared (R)
r2 = r2_score(y_test, final_predictions)
print(f"R-squared (R): {r2}")
```

```python
end_time = time.time()

# Calculate execution time for MapReduce
mapreduce_time = end_time - start_time
print(f"MapReduce Model Execution Time: {mapreduce_time:.4f} seconds")
```
